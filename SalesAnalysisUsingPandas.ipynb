# What Products are often Sold Together

```python
df=all_data[all_data['Order ID'].duplicated(keep=False)]
df['ItemsBoughtTogether']=all_data.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
df=df[['Order ID','ItemsBoughtTogether']].drop_duplicates()
df.head(20)
```

```python
# Items which are frequently bought together

from itertools import combinations
from collections import Counter
count=Counter()
for row in df['ItemsBoughtTogether']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2))) 
#count.most_common(10)
for key,value in count.most_common(10):
    print(key,value)
```

# What Product sold the most and why do you think it sold the most

```python
# Which product sold the most and why do you think so!

# How t
Product_Groupby=all_data.groupby('Product')
quantity_ordered=Product_Groupby.sum()['Quantity Ordered']
products=[product for product, df in Product_Groupby]
m.bar(products,quantity_ordered)
m.xticks(products,rotation='vertical',size=8)
m.ylabel('Number of Orders')
m.xlabel('Product Name')
m.show()
```

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']

keys = [pair for pair, df in product_group]
m.bar(keys, quantity_ordered)
m.xticks(keys, rotation='vertical', size=8)
m.show()
```

```python
prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = m.subplots()

ax2 = ax1.twinx()
ax1.bar(keys, quantity_ordered, color='g')
ax2.plot(keys, prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(keys, rotation='vertical', size=8)

fig.show()
```
